TENSOR INDEXING

```python
tensor_new=tf.constant([1,2,3,4,5,6]) ##0 till n
print(tensor_new)
print(tensor_new[0:3]) ## min:max-1
print(tensor_new[:3])
print(tensor_new[0:])
print(tensor_new[0:-3]) # n-1=-1
```

```python
tensor_two_d=tf.constant([[1,2,0],
                          [3,5,-1],
                           [1,5,6],
                          [2,3,8]])
print(tensor_two_d)
print(tensor_two_d[0:3,1:3])  ## row(min:max+1),col(min:max+1)
print(tensor_two_d[0:3,:])
print(tensor_two_d[2,:])
print(tensor_two_d[2,0])
print(tensor_two_d[...,1]) ##...=>complete row
```

```python
tensor_three_d=tf.constant([
  [[1,2,0],
  [3,5,-1]],
  [[1,2,0],
  [3,5,-1]]
  ])
print(tensor_three_d)
print(tensor_three_d[0 , : , :])
```

MATHS OPERATIONS

```python
#tf.math
#abs
x_abs=tf.constant([-10,-10,30])
tf.math.abs(x_abs)
tf.abs(x_abs)

x_abs_comples=tf.constant([-2.1+0.7j])
print(tf.abs(x_abs_comples))  ##root of a*a+b*b
```

```python
#add  , subtract , multiply, divide (elementwise) of same shape bith tensors
x_1=tf.constant([1,2,3])
x_2=tf.constant([4,5,6])
print(tf.add(x_1,x_2))
print(tf.math.add(x_1,x_2))
print(tf.math.subtract(x_1,x_2))
print(tf.math.multiply(x_1,x_2))
print(tf.math.divide(x_1,x_2))

#divide_no_nan of same shape
x_1=tf.constant([1,2,3])
x_2=tf.constant([0,2,3])
print(tf.math.divide_no_nan(x_1,x_2))
print(tf.math.divide(x_1,x_2))
```

```python
#add  , subtract , multiply, divide (elementwise) of diff shape tensors
x_1=tf.constant([1,2,3])
x_2=tf.constant([4])
print(tf.add(x_1,x_2))        ###====== BROADCASTING X_2 STRECHED TO SAME DIMENSION AS OF X_1=====
print(tf.math.add(x_1,x_2))
print(tf.math.subtract(x_1,x_2))
print(tf.math.multiply(x_1,x_2))
print(tf.math.divide(x_1,x_2))

#divide_no_nan of same shape
x_1=tf.constant([1,2,3])
x_2=tf.constant([0])
print(tf.math.divide_no_nan(x_1,x_2))
print(tf.math.divide(x_1,x_2))
```

```python
# BROADCASTING ANOTHER EG
x1=tf.constant([5,3,6,6,4,6])
x2=tf.constant([[7],[5],[3]])
print(x1.shape)
print(x2.shape)
print(tf.math.multiply(x1,x2))
```

```python
#argmax  => gives index of max element
x_argmax=tf.constant([10,20,30,40])
print(tf.math.argmax(x_argmax))


multidim=tf.constant([   # for multidim tensor needs to fix the axis
    [2,20,30,3,6],
    [3,11,16,1,8],
    [144,45,23,5,27]
])
print(tf.math.argmax(multidim,0))   # fix rows(0) comparison along col1
print(tf.math.argmax(multidim,1))
```

```python
#pow
x1= tf.constant([[2,2],[3,3]])
x2=tf.constant([[1,2],[3,4]])
print(tf.math.pow(x1,x2))
```
